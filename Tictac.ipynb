```python
#Tictak toe
def display(lst):
    i_lst=0
    for i in range (5):
        if i%2!=0:
            print("---------")
        else:
            for j in range (5):
                if j%2==0:
                    print(lst[i_lst],end="")
                    i_lst+=1
                else:
                    print(" | ",end="")
            print()
def diagonal_check(lst):
    count_check1=0
    count1=0
    for index1 in range (3):
        if lst[index1]=='X':
            count_check1+=1
            if count_check1==3:
                return True
                break
        elif lst[index1]=='O':
            count1+=1
            if count1==3:
                return True
                break
    count_check1=0
    count1=0
    for index2 in range (3,6):
        if lst[index2]=='X':
            count_check1+=1
            if count_check1==3:
                return True
        elif lst[index2]=='O':
            count1+=1
            if count1==3:
                return True
    count_check1=0
    count1=0
    for index3 in range (6,9):
        if lst[index3]=='X':
            count_check1+=1
            if count_check1==3:
                return True
        elif lst[index3]=='O':
            count1+=1
            if count1==3:
                return True
def Cross_check(lst):
    count_check2=0
    count2=0
    for index4 in range (0,9,4):
        if lst[index4]=='X':
            count_check2+=1
            if count_check2==3:
                return True
        elif lst[index4]=='O':
            count2+=1
            if count2==3:
                return True
    count_check2=0
    count2=0
    for index5 in range (2,7,2):
        if lst[index5]=='X':
            count_check2+=1
            if count_check2==3:
                return True
        elif lst[index5]=='O':
            count2+=1
            if count2==3:
                return True
def coloum_check(lst):
    count_check3=0
    count3=0
    for index6 in range (0,7,3):
        if lst[index6]=='X':
            count_check3+=1
            if count_check3==3:
                return True
        elif lst[index6]=='O':
            count3+=1
            if count3==3:
                return True
    count_check3=0
    count3=0
    for index7 in range (1,9,3):
        if lst[index7]=='X':
            count_check3+=1
            if count_check3==3:
                return True
        elif lst[index7]=='O':
            count3+=1
            if count3==3:
                return True
    count_check3=0
    count3=0
    for index8 in range (2,9,3):
        if lst[index8]=='X':
            count_check3+=1
            if count_check3==3:
                return True
        elif lst[index8]=='O':
            count3+=1
            if count3==3:
                return True
def check(lst):
    if diagonal_check(lst):
            return True
    elif Cross_check(lst):
            return True
    elif coloum_check(lst):
            return True
def isDraw(lst):
    count4=0
    for index9 in range (9):
        if lst[index9]=='X' or lst[index9]=='O':
            count4+=1
    if count4==9:
        return True
lst=['1','2','3','4','5','6','7','8','9']
selectplayer=1
display(lst)
n=0
while True:
    if selectplayer==1:
        n=str(input("Player one:"))
    else:
        n=str(input("Player two:"))
    if n=="X" or n=="O":
        print("Invalid input please try to enter a number from 1 to",len(lst))
        continue
    if n not in lst:
        print("Invalid input please try to enter a number from 1 to",len(lst))
        continue
    n=int(n)
    if selectplayer==1:
        lst[n-1]='X'
    else:
        lst[n-1]='O'
    display(lst)
    if check(lst):
        print("Player "+ str(selectplayer) + " wins")
        break
    if isDraw(lst):
        print("match draw")
        break
    if selectplayer==1:
        selectplayer=2
    else:
        selectplayer=1
```
